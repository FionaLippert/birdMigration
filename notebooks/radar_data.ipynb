```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/home/fiona/birdMigration/data/preprocessed/radar/fall/2016/dynamic_features.csv')
```

```python
radars = df.radar.unique()
fig, ax = plt.subplots(radars.size, 1, figsize=(10, 3*radars.size))
for idx, r in enumerate(radars):
    df[df.radar == r].birds.plot(ax=ax[idx])
```
